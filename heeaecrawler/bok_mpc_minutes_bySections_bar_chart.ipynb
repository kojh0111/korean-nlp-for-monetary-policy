```python
import matplotlib.pyplot as plt
import pandas as pd

data_df = pd.read_csv("C:/Users/kwkwo/sesac_studygroup/bok_mpc_minutes_into_sections_data.csv")

#plt.rc('font', family = 'AppleGothic') # mac 
plt.rc('font', family = 'Malgun Gothic') # window
plt.rc('font', size = 12)
plt.rc('axes', unicode_minus = False) # -표시 오류 잡아줌
```

```python
data_df
```

```python
# Group the data by 'year' and 'section', and sum the 'sent-no' values
df_grouped = data_df.groupby(['year', 'section'])['sent-no'].mean().unstack(fill_value=0)

# Create a stacked bar plot
df_grouped.plot(kind='bar', stacked=True, figsize=(10, 6))

# Add title and labels
plt.title('Cumulative Sentence Counts by Section and Year')
plt.xlabel('Year')
plt.ylabel('Average no. of Sentences')

# Display legend
plt.legend(title='Section')

# Show plot
plt.tight_layout()
plt.show()
```
